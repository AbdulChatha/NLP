```python
import pandas as pd
import numpy as np
```

```python
# data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/twitter.csv")
data=pd.DataFrame({"tweet": [
    "Galactic quest thrills, pacing stumbles stumbles.",
    "Stars explored, mesmerizing cosmic voyage.",
    "Nebula Adventure: cosmic roller coaster!"
]
})
```

## Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer( ) #to make n grams set ngram_range=(n,m) e.g ngram_range=(2,2) which is bigram
#Bag of words similar to unigrams
```

```python
BOW=cv.fit_transform(data['tweet'])
BOW
```

```python
cv.vocabulary_
```

```python
print(cv.get_feature_names_out())
print(BOW.toarray())
```
